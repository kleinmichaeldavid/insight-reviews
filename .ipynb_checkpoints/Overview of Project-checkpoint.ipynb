## Sentiment Model

```python
## process the data for the sentiment model
## trained on full reviews (not split into sentences)
## sklearn tfidf vectorizer takes full sentences (not tokenized)
## leave in stop words so they can be included in bigrams

data_for_sentiment = preprocessing(data,
                                   split_sentences = False,
                                   replace_numbers = True,
                                   tokenize = False,
                                   lemmatize = True,
                                   remove_stops = False,
                                   embed = False)
```

```python
## training and validation set
X_train, X_test, y_train, y_test = train_test_split(data_for_sentiment['clean_text'],
                                                    data_for_sentiment['y'],
                                                    test_size=0.25,
                                                    random_state=20313119)
```

```python
## oversample negative reviews
ros = RandomOverSampler(random_state=0)
X_resampled, y_resampled = ros.fit_resample(X_train.to_frame(), y_train)
X_resampled = pd.Series(X_resampled[:,0]); y_resampled = pd.Series(y_resampled)
```

```python
model_nb = make_pipeline(TfidfVectorizer(ngram_range = (1,2)),
                         MultinomialNB())

model_nb.fit(X_resampled, y_resampled)
```

```python
## train a version of the model on the full dataset (train + val) to be used in the web app
model_full = make_pipeline(TfidfVectorizer(ngram_range = (1,2)),
                           MultinomialNB())

model_full.fit(data_for_sentiment['clean_text'],
               data_for_sentiment['y'])

joblib.dump(model_full, 'model_nb_200sellers.pkl')
```

### Comparison to VADER model

```python
analyser = SentimentIntensityAnalyzer()

## initialize arrays for probabilities and sentiments
probabilities = np.zeros(X_test.shape[0])
sentiments = np.repeat(True, X_test.shape[0])

## run vader on each review
for i,review in enumerate(X_test):
    snt = analyser.polarity_scores(review)
    neg = snt['neg']
    pos = snt['pos']
    neu = snt['neu']
    
    # for neutral sentiment, put half toward positive
    pos_prob = pos + neu / 2
    probabilities[i] = pos_prob
    if pos_prob > 0.5:
        sentiments[i] = True
    else:
        sentiments[i] = False
```

```python
# Plot normalized confusion matrix for validation data
plot_confusion_matrix(y_test, sentiments,
                      classes=np.array(['Negative','Positive']),
                      normalize=True,
                      title='VADER Sentiment Analysis - Validation Set')

plt.show()
```

```python
## ROC curve for VADER
roc_auc_vader = roc_auc_score(y_test, probabilities)
fpr_vader, tpr_vader, thresholds = roc_curve(y_test, probabilities)
```

```python
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='green',
         lw=lw, label='ROC curve (Naive Bayes) (area = %0.2f)' % roc_auc)
plt.plot(fpr_vader, tpr_vader, color='darkorange',
         lw=lw, label='ROC curve (VADER baseline) (area = %0.2f)' % roc_auc_vader)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Sentiment Analysis of Full Reviews')
plt.legend(loc="lower right")
plt.show()
```

### Check important features

```python
## extract an array of the features
features = model_nb['tfidfvectorizer'].get_feature_names()

## probability of each feature for each class
pos_class_prob = np.exp(model_nb['multinomialnb'].feature_log_prob_[1,:])
neg_class_prob = np.exp(model_nb['multinomialnb'].feature_log_prob_[0,:])

## probability of positive based on token being in review
pos_probability = pos_class_prob / (pos_class_prob + neg_class_prob)
pos_probability_sorted = pos_probability.argsort()
```

```python
plt.barh(np.take(features, pos_probability_sorted[:20]),
         np.take(pos_probability, pos_probability_sorted[:20]))
plt.xlim([0,1])
plt.title('Most Negative Features')
plt.xlabel('Probability of Review Being Positive')
```

```python
plt.barh(np.take(features, pos_probability_sorted[-20:]),
         np.take(pos_probability, pos_probability_sorted[-20:]))
plt.xlim([0,1])
plt.title('Most Positive Features')
plt.xlabel('Probability of Review Being Positive')
```
